## 슬픔,상처 비교

```python
# 기본적인 통계 비교
hurt_texts = df[df['감정_대분류'] == '상처']['사람문장1']
sad_texts = df[df['감정_대분류'] == '슬픔']['사람문장1']

print(f"'상처' 텍스트 평균 길이: {hurt_texts.str.len().mean():.2f}")
print(f"'슬픔' 텍스트 평균 길이: {sad_texts.str.len().mean():.2f}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from konlpy.tag import Okt

okt = Okt()

# 단어 추출 함수
def get_words(texts):
    all_words = []
    for text in texts:
        # 명사 추출
        nouns = okt.nouns(text)
        
        # 형용사 추출 - pos 함수를 사용하여 품사 태깅 후 형용사(Adjective) 필터링
        pos_tagged = okt.pos(text)
        adjectives = [word for word, pos in pos_tagged if pos == 'Adjective']
        
        # 명사와 형용사 합치기
        words = nouns + adjectives
        all_words.extend(words)
    
    return ' '.join(all_words)

# 워드클라우드 생성
hurt_texts = df[df['감정_대분류'] == '상처']['사람문장1']
sad_texts = df[df['감정_대분류'] == '슬픔']['사람문장1']

hurt_words = get_words(hurt_texts)
sad_words = get_words(sad_texts)

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# 상처 워드클라우드
wc_hurt = WordCloud(font_path='/System/Library/Fonts/AppleGothic.ttf', width=800, height=600, 
                   background_color='white').generate(hurt_words)
axes[0].imshow(wc_hurt)
axes[0].set_title('상처 관련 단어', fontsize=20)
axes[0].axis('off')

# 슬픔 워드클라우드  
wc_sad = WordCloud(font_path='/System/Library/Fonts/AppleGothic.ttf', width=800, height=600, 
                  background_color='white').generate(sad_words)
axes[1].imshow(wc_sad)
axes[1].set_title('슬픔 관련 단어', fontsize=20)
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# 상위 20개 단어 비교
hurt_counter = Counter(hurt_words.split())
sad_counter = Counter(sad_words.split())

print("'상처' 상위 20개 단어:")
print(hurt_counter.most_common(20))

print("\n'슬픔' 상위 20개 단어:")
print(sad_counter.most_common(20))

# 공통 단어와 고유 단어 비교
hurt_top50 = set(dict(hurt_counter.most_common(50)).keys())
sad_top50 = set(dict(sad_counter.most_common(50)).keys())

common_words = hurt_top50.intersection(sad_top50)
print(f"\n공통 단어(상위 50개 중): {common_words}")
print(f"'상처'에만 있는 단어: {hurt_top50 - sad_top50}")
print(f"'슬픔'에만 있는 단어: {sad_top50 - hurt_top50}")
```

```python
# 1. 각 문장 유형별로 데이터 추출 및 재구성
def restructure_data(df):
    # 문장1 데이터 추출
    sentences1 = df[['감정_대분류', '사람문장1']].copy()
    sentences1['문장_유형'] = '사람문장1'
    sentences1 = sentences1.rename(columns={'사람문장1': '문장내용'})
    
    # 문장2 데이터 추출
    sentences2 = df[['감정_대분류', '사람문장2']].copy()
    sentences2['문장_유형'] = '사람문장2'
    sentences2 = sentences2.rename(columns={'사람문장2': '문장내용'})
    
    # 문장3 데이터 추출
    sentences3 = df[['감정_대분류', '사람문장3']].copy()
    sentences3['문장_유형'] = '사람문장3'
    sentences3 = sentences3.rename(columns={'사람문장3': '문장내용'})
    
    # 세 데이터프레임 수직으로 합치기
    combined = pd.concat([sentences1, sentences2, sentences3], ignore_index=True)
    
    # NaN 값 제거
    combined = combined.dropna()
    
    return combined

# 데이터 재구성
restructured_df = restructure_data(df)

restructured_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# 상처와 슬픔 데이터만 추출
emotion_df = restructured_df[restructured_df['감정_대분류'].isin(['상처', '슬픔'])]
X = emotion_df['문장내용']
y = emotion_df['감정_대분류']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# TF-IDF 벡터화
tfidf = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# 모델 학습
model = LogisticRegression(random_state=42)
model.fit(X_train_tfidf, y_train)

# 예측 및 평가
y_pred = model.predict(X_test_tfidf)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import random
import re
import pandas as pd

# 불필요한 감정 표현 제거 패턴 (ㅋㅋ, ㅎㅎ, ㅠㅠ, ~, 이모지 등)
filter_pattern = re.compile(r"[ㅋㅎㅠㅜ~^…]+|[^\w\s가-힣.,?!]")

# 대화 패턴
patterns = {
    "진술": [
        "{}에 다녀왔어.", "{}에서 일해.", "{}을 했어.", "{}이 맛있었어.", "{}가 필요해."
    ],
    "질문": [
        "{} 먹어봤어?", "{} 어디 있는지 알아?", "{} 어떻게 생각해?", "{} 지금 해도 돼?", "{} 알아?"
    ],
    "명령": [
        "{} 해줘.", "{} 좀 챙겨줘.", "{} 정리해줘.", "{} 알려줘.", "{} 닫아줘."
    ],
    "정보": [
        "{}은 {}야.", "{}는 {}에서 만들어졌어.", "{}는 {}에 있어.", "{}의 특징은 {}야."
    ]
}

# 채워넣을 단어들
fillers = {
    "장소": ["회사", "집", "학교", "마트", "도서관", "카페", "식당", "병원"],
    "물건": ["문", "창문", "컴퓨터", "서류", "이메일", "전화기", "책"],
    "음식": ["김밥", "라면", "커피", "과일", "떡볶이", "샌드위치"],
    "대상": ["서울", "지구", "컴퓨터", "인터넷", "고양이", "물", "전기"],
    "속성": ["수도", "행성", "기계", "전자기기", "포유류", "에너지"]
}

# 문장 생성 함수
def generate_sentence():
    category = random.choice(list(patterns.keys()))
    template = random.choice(patterns[category])
    if "{}은 {}야." in template or "{}는 {}에 있어." in template or "{}는 {}에서 만들어졌어." in template or "{}의 특징은 {}야." in template:
        return template.format(random.choice(fillers["대상"]), random.choice(fillers["속성"]))
    else:
        word = random.choice(
            fillers["장소"] + fillers["물건"] + fillers["음식"] + fillers["대상"]
        )
        return template.format(word)

# 감정 표현 없는 문장 9000개 생성
neutral_sentences = []
while len(neutral_sentences) < 9000:
    sentence = generate_sentence()
    if not filter_pattern.search(sentence):
        neutral_sentences.append(sentence)

# 데이터프레임으로 저장
df = pd.DataFrame({"text": neutral_sentences, "emotion": "none"})
df.to_csv("neutral_conversations_9000.csv", index=False)
print("✅ 9,000개 중립 대화문 생성 완료 및 저장됨.")
```

```python
import json
import pandas as pd
import os
from glob import glob
from tqdm import tqdm
import re

# 기준 폴더 경로
base_dir = "/Users/hwangeunbi/chatnge_AI/공감형 대화/01-1.정식개방데이터/Training/02.라벨링데이터"

def empathy_sentences():
    # 결과를 저장할 리스트 초기화
    모든_대화_쌍 = []
    메타데이터_목록 = []
    파일_경로_목록 = []
    
    # 모든 JSON 파일 찾기
    json_파일목록 = []
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.endswith(".json"):
                json_파일목록.append(os.path.join(root, file))
    
    print(f"총 {len(json_파일목록)}개의 JSON 파일을 찾았습니다.")
    
    # 모든 JSON 파일 처리
    for 파일_경로 in tqdm(json_파일목록, desc="파일 처리 중"):
        try:
            # JSON 파일 로드
            with open(파일_경로, 'r', encoding='utf-8') as f:
                데이터 = json.load(f)
            
            # 폴더 경로에서 카테고리 정보 추출 (선택사항)
            폴더_정보 = 파일_경로.split(base_dir)[1]
            카테고리_패턴 = r"TL_(.+?)_"
            감정_카테고리 = re.search(카테고리_패턴, 폴더_정보)
            감정_카테고리 = 감정_카테고리.group(1) if 감정_카테고리 else None
            
            # 메타데이터 추출
            메타데이터 = {
                'id': 데이터['info']['id'],
                '카테고리': 데이터['info']['category'],
                '상황': 데이터['info']['situation'],
                '화자_감정': 데이터['info']['speaker_emotion'],
                '관계': 데이터['info']['relation'],
                '청자_행동': 데이터['info'].get('listener_behavior', []),
                '평균_평점': 데이터['info']['evaluation'].get('avg_rating'),
                '파일_경로': 파일_경로,
                '추출_감정_카테고리': 감정_카테고리
            }
            메타데이터_목록.append(메타데이터)
            파일_경로_목록.append(파일_경로)
            
            # 대화 쌍 추출 (스피커-리스너 순서로)
            발화목록 = 데이터['utterances']
            for i in range(1, len(발화목록)):
                if 발화목록[i]['role'] == 'listener' and 발화목록[i-1]['role'] == 'speaker':
                    대화_쌍 = {
                        '대화_id': 데이터['info']['id'],
                        '맥락': 발화목록[i-1]['text'],
                        '응답': 발화목록[i]['text'],
                        '공감_유형': 발화목록[i]['listener_empathy'],
                        '맥락_id': 발화목록[i-1]['utterance_id'],
                        '응답_id': 발화목록[i]['utterance_id'],
                        '상황': 데이터['info']['situation'],
                        '화자_감정': 데이터['info']['speaker_emotion'],
                        '관계': 데이터['info']['relation'],
                        '파일_경로': 파일_경로,
                        '추출_감정_카테고리': 감정_카테고리
                    }
                    모든_대화_쌍.append(대화_쌍)
        
        except Exception as e:
            print(f"파일 처리 중 오류 발생: {파일_경로}")
            print(f"오류 내용: {e}")
    
    # 데이터프레임 생성
    메타데이터_df = pd.DataFrame(메타데이터_목록)
    대화_쌍_df = pd.DataFrame(모든_대화_쌍)
    
    # 공감 유형 분석 (None 값 처리)
    대화_쌍_df['공감_유형'] = 대화_쌍_df['공감_유형'].apply(lambda x: [] if x is None else x)
    
    # 공감 유형별 대화 쌍 개수 계산
    공감_유형_통계 = {}
    for 인덱스, 행 in 대화_쌍_df.iterrows():
        유형목록 = 행['공감_유형']
        if not 유형목록:
            공감_유형_통계['없음'] = 공감_유형_통계.get('없음', 0) + 1
        else:
            for 유형 in 유형목록:
                공감_유형_통계[유형] = 공감_유형_통계.get(유형, 0) + 1
    
    print("\n== 데이터 통계 ==")
    print(f"총 대화 파일 수: {len(메타데이터_목록)}")
    print(f"총 대화 쌍 수: {len(모든_대화_쌍)}")
    print("\n== 공감 유형별 개수 ==")
    for 유형, 개수 in 공감_유형_통계.items():
        print(f"{유형}: {개수}")
    
    # GPT-4 학습을 위한 데이터 형식 생성
    gpt_학습_데이터 = []
    
    for i, 행 in 대화_쌍_df.iterrows():
        # 공감 유형 문자열로 변환
        공감_유형_문자열 = ", ".join(행['공감_유형']) if 행['공감_유형'] else ""
        
        # 시스템 메시지 생성
        시스템_메시지 = f"""당신은 공감 능력이 뛰어난 대화 AI입니다. 
상황: {행['상황']}
화자의 감정: {행['화자_감정']}
대화 관계: {행['관계']}
공감 유형: {공감_유형_문자열}"""
        
        # 대화 쌍 생성
        대화 = {
            "messages": [
                {"role": "system", "content": 시스템_메시지},
                {"role": "user", "content": 행['맥락']},
                {"role": "assistant", "content": 행['응답']}
            ]
        }
        
        gpt_학습_데이터.append(대화)
    
    return 메타데이터_df, 대화_쌍_df, gpt_학습_데이터

# 함수 실행
메타데이터_df, 대화_쌍_df, gpt_학습_데이터 = empathy_sentences()

# 결과 저장
메타데이터_df.to_csv('공감대화_메타데이터.csv', index=False, encoding='utf-8-sig')
대화_쌍_df.to_csv('공감대화_대화쌍.csv', index=False, encoding='utf-8-sig')

# GPT-4 학습 데이터 저장 (JSONL 형식)
with open('공감대화_gpt_학습데이터.jsonl', 'w', encoding='utf-8') as f:
    for 항목 in gpt_학습_데이터:
        f.write(json.dumps(항목, ensure_ascii=False) + '\n')

print("\n✅ 모든 데이터 전처리 완료!")
```
